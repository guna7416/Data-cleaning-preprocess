```python
#boxplots for outliners
sns.boxplot(x=df['Fare'])
plt.title('Fare Distribution')
plt.show()
```

```python
#countplots for categories
sns.countplot(x='Survived', data=df)
plt.title('Survival Count')
plt.show()
```

```python
#correlation matrix
# Only select numeric columns for correlation
corr = df.corr(numeric_only=True)

plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

###  Observation

- `Sex` and `Pclass` are highly correlated with `Survived`.
- We will use these features for model training.


```python
#pair plot
sns.pairplot(df[['Survived', 'Age', 'Fare', 'Pclass']].dropna(), hue='Survived')
plt.show()
```

###  Observation

- The correlation matrix shows that `Pclass`, `Sex`, and `Fare` are strongly related to `Survived`.
- `Name`, `Ticket`, and `Cabin` seem unrelated and can be dropped.
- We will select the most relevant features for training our model.
